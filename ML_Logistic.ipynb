<h1> Логистическая регрессия

Логистическая регрессия -- линейный классификатор, использующий логарифмическую функцию потерь. Чтобы сформулировать постановку задачи логистической регрессии сначала разберем более общую задачу.

 $X$ — множество объектов  
 $Y$ — множество ответов  
 $y: X\to Y$ — неизвестная зависимость (target function)  
 Дано: обучающая выборка — $(x_1,\ldots, x_n) \subset X$, $y_i=y(x_i), ~i=1,\ldots,n$ — известные ответы.  
 Найти: $a:X\to Y$ — функцию (decision function), приближающую $y$ на всем множестве $X$.  

Задача классификации с двумя классами, $Y=\{-1,+1\}$:  
по обучающей выборке $X^n=(x_i,y_i)_{i=1}^n$ построить алгоритм классификации $a(x,w)=\textrm{sign} f(x,w)$,  
где $f(x,w)$ — разделяющая (дискриминантная) функция, $w$ — вектор параметров.  
    
$f(x,w)=0$ — разделяющая поверхность;  
 
$M_i(w)=y_if(x_i,w)$ — отступ (margin) объекта $x_i$;  
    
$M_i(w)<0 \Leftrightarrow$  алгоритм $a(x,w)$ ошибается на $x_i$.  

Задача: минимизировать число ошибок классификации 	 
\begin{equation*}
Q(w)=\sum \limits_{i=1}^n [M_i(w)<0]\le \tilde Q(w)=\sum \limits_{i=1}^n \mathcal{L}(M_i(w)) \to \min \limits_{w}.
\end{equation*}

### Линейный классификатор: 
$a(x,w)=\text{sign}~ (\sum\limits_{j=1}^pw_jx^{j}-w_0)$, $w_0,w_1,\ldots,w_p\in\mathbb{R}$.

Пусть $x^0=-1$, тогда

 $$a(x,w)=\text{sign}~\langle w,x\rangle ,~ x,w\in\mathbb{R}^{p+1}.$$
 
 
 $M_i(w)=y_i\langle w,x_i\rangle$ — отступ объекта $x_i$. 
 
Задача: $$Q(w)=\sum \limits_{i=1}^n [y_i\langle w,x_i\rangle <0] \le \sum \limits_{i=1}^n \mathcal{L}(y_i\langle w,x_i\rangle) \to \min \limits_{w}$$

$$Q(w)=\alpha_{1}\sum_{i\in I_{1}}[y_i\neq \hat{y_{i}}]+\alpha_{2}\sum_{i\in I_{2}}[y_i\neq \hat{y_{i}}]\leq
\alpha_{1}\sum_{i\in I_1}\mathcal{L}(y_{i}\langle w,x_i\rangle)+\alpha_{2}\sum_{i\in I_2}\mathcal{L}(y_{i}\langle w,x_i\rangle)$$

### Подход через минимизацию функции потерь

Линейная модель классификации: $a(x)=\text{sign}~ \langle w,x\rangle,~~~ x,w\in \mathbb{R}^{p+1},~~ M_i(x)=\langle w,x_{i} \rangle y_{i}$ --- отступ. 

В качестве аппроксимации пороговой функции потерь берется логарифмическая функция потерь $\mathcal{L}(M)=\log(1+e^{-M})$ и задача выглядит следующим образом: 
$$\tilde Q(w)=\sum \limits_{i=1}^n \log \left(1+\exp{(-y_i\langle w,x_i\rangle)}\right) \to \min \limits_{w}.$$

Способы решения задачи минимизации:  
* метод стохастического градиента, 
* метод Ньютона-Рафсона.

**Метод стохастического градиента**. 

$\tilde Q(w)=\sum_{i=1}^{n}\mathcal{L}(w,x_{i},y_{i})\to\min_{w}.$

Вход: $\mathbb{X}_n=(x_i,y_i)_{i=1}^n, h, \lambda$   
Выход: $w$  

* Инициализация $w_j,~ j=0,\ldots,p$ 
* Инициализация $\tilde Q(w)$ 
* Повторять  
** Выбор $x_i$ из $\mathbb{X}_n$ случайным образом  
** Вычисление $\varepsilon_i := \mathcal{L}_i(w)$  
** Градиентный шаг $w:=w-h\nabla\mathcal{L}_i(w)$  
** Вычисление $\tilde Q(w):=\lambda\varepsilon_i+(1-\lambda)\tilde Q(w)$ 	 
	пока $\tilde Q(w)$ или $w$ не сойдутся


### Регуляризация  
$L2$-регуляризация:
\begin{equation*}
\tilde Q_{C}(w)=\tilde Q(w)+\frac{1}{2C}\sum_{j=1}^{p}w_{j}^{2}\to \min \limits_w.
\end{equation*}
$L1$-регуляризация:
\begin{equation*}
\tilde Q_{C}(w)=\tilde Q(w)+\frac{1}{C}\sum_{j=1}^{p}|w_{j}|\to \min \limits_w.
\end{equation*}

### Вероятностный подход

Хотим оценить вероятности принадлежности какому-то классу.

Пусть $Y =\{-1,1\}.$ Пусть $P(y=-1|x,w) = \sigma_w(x)= \frac{1}{1+e^{\langle w,x\rangle}}.$ (Делается предположение о том, что вероятность наступления события y=-1 такова) 

Тогда вероятность принять значение 1 равна: $P(y=1|x,w)=1-\sigma_w(x)=\sigma_w(-x).$

Тогда $P(y=y_{i}|x_i;w)=\sigma_w(-y_ix_i)=\frac{1}{1+e^{-y_i\langle w,x\rangle}}.$

Выписываем функцию правдоподобия: 
$$Q(w)=-\log L(w)=-\log \prod\limits_{i=1}^nP(y=y_i|x_i,w)=-\sum_{i=1}^{n}\log P(y=y_i|x_i,w)\to\min_{w} $$
    
$-\log P(y=y_i|x_i,w)=-\log\frac{1}{1+e^{-y_i\langle w,x\rangle}}=\log(1+e^{-M})$, где $M=y_{i}\langle w,x\rangle$ -- отступ.

## LogisticRegression
 class sklearn.linear_model.LogisticRegression(penalty=’l2’,tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1,class_weight=None, random_state=None, solver=’liblinear’,max_iter=100)
	
* penalty: ’l1’, ’l2’(default = ’l2’),– вид регуляризации, “lbfgs”и “newton-cg” поддреживают только L2 регуляризвцию.
* C: float(default=1.0)– параметр регуляризации, должен быть положительным float. Меньшее С обеспечивает более сильную регуляризацию.
* fit_intercept : bool(default = True) – определяет добавлять ли константу к решающей функции.
* intercept_scaling : float(default = 1) – полезен, только если используется libliner, если self.fit_intercept=True, то векторx, описывающий объект, становится равным [x,self.intercept_scaling].
* class_weight : dict or ‘balanced’(default=None)– веса, ассоциированные с классами, записываются в форме{class_label: weight}. Если параметр не указан, все веса предполагаются равными 1. ’balanced’ использует значения y, чтобы автоматически установить веса обратнопропорционально частотам класса во входных данных.
* random_state: int seed, RandomState instance, or None(default) – для инициализации генератора псевдослучайных чисел при перемешивании данных.
* solver : ‘newton-cg’, ‘lbfgs’, ‘liblinear’, ‘sag’(default =’liblinear’), – алгоритм, который используется для решения проблемы оптимизации. ('sag' -- Stochastic Average Gradient)
* max_iter : int(default=100) – используется только для newton-cg, sag и lbfgs. Максимальное число итераций.
* multiclass: ’ovr’, ’multinimial’. При использовании ’ovr’ используется стратегия OneVsRest для мультиклассовойклассификации. Иначе, минимизируемая функция потерь это multinomial loss, которая учитывает полное распределение вероятностей.

Логистическая регрессия - это линейный алгоритм (с нелинейным преобразованием на выходе). Он предполагает линейную связь между входными переменными и выходными данными. Преобразования данных ваших входных переменных, которые лучше раскрывают эти линейные отношения, могут привести к более точной модели. Наши признаки нужно логарифмировать. Отправим логарифмироваться все признаки с большим эксцессом или ассиметрией. (это все на самом деле, оно и видно)

```python
numeric_feats = numerical_columns
skewed_feats = X_train[numeric_feats].apply(lambda x: skew(x.dropna()))
kurtosis_feats = X_train[numeric_feats].apply(lambda x: kurtosis(x.dropna()))
skewed_feats = skewed_feats[skewed_feats > 0.75]
kurtosis_feats = kurtosis_feats[kurtosis_feats > 0.75]
kurtosis_feats = kurtosis_feats.index
skewed_feats = skewed_feats.index
feats = list(set(list(kurtosis_feats) + list(skewed_feats)))

X_train[feats] = np.log1p(X_train[feats]+1)
X_test[feats] = np.log1p(X_test[feats]+1)
```

```python
dataf=(X_train).copy()
dataf['class']=y_train
```

```python
def Print_Classification_report05(estimator, X_train, y_train, X_test, y_test,threshold=0.5):
    logreg = estimator
    logreg.fit(X_train, y_train)
    score_train = logreg.score(X_train, y_train)
    score_test = logreg.score(X_test, y_test)
    predictionsLOGIT = logreg.predict_proba(X_test)[:,1]
    roc_auc=roc_auc_score(y_test,predictionsLOGIT)
    y_pred = np.int32(np.zeros(len(predictionsLOGIT)))
    for i in range(len(y_pred)):
        y_pred[i] = predictionsLOGIT[i] > threshold
    conf_mat = metrics.confusion_matrix(y_test, y_pred)
    conf_mat = pd.DataFrame(conf_mat)
    print("Confusion matrix: ")
    print(conf_mat)
    print("Classification report: ")
    print(classification_report(y_test, y_pred,digits=5))
```
